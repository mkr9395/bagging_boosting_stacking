Ensemble learning is a technique in machine learning where multiple models (often called *base learners* or *weak learners*) are combined to solve a particular task and improve performance compared to individual models.

### 🔰 **Main Types of Ensemble Learning**

---

### 1. **Bagging (Bootstrap Aggregating)**

* **Goal:** Reduce **variance** and avoid overfitting.
* **How it works:**

  * Multiple same models are trained on different **random subsets** of the training data (sampled with replacement).
  * Predictions are **averaged** (for regression) or **voted** (for classification).
* **Example Algorithms:**

  * **Random Forest**
* **Use Case:** When the base model is **high variance** (e.g., decision trees).

---

### 2. **Boosting**

* **Goal:** Reduce **bias** and build a strong learner by sequentially learning from errors.
* **How it works:**

  * Models are trained **sequentially**, each focusing on the **errors** made by the previous one.
  * Final prediction is a **weighted sum** of all models.
* **Example Algorithms:**

  * **AdaBoost**
  * **Gradient Boosting Machines (GBM)**
  * **XGBoost**
  * **LightGBM**
  * **CatBoost**
* **Use Case:** When you want to improve **accuracy** by correcting weak predictions.

---

### 3. **Stacking (Stacked Generalization)**

* **Goal:** Improve predictions by **learning how to combine** multiple models.
* **How it works:**

  * Multiple base models are trained.
  * Their predictions are then used as input features for a **meta-model** (blender) that learns the best way to combine them.
* **Example:** Combining logistic regression, decision tree, and SVM outputs with a meta-model like a random forest.
* **Use Case:** When you want to leverage the **strengths of diverse models**.

---

### 4. **Voting**

* **Goal:** Combine multiple models and use majority or average to decide.
* **How it works:**

  * **Hard Voting:** Each model votes, and the majority class is chosen.
  * **Soft Voting:** Each model provides class probabilities, and average probabilities are used.
* **Example:** VotingClassifier in scikit-learn
* **Use Case:** Simple yet effective way to **aggregate model predictions**.

---

### 5. **Blending**

* Similar to stacking but simpler:

  * Use a **holdout validation set** to train the meta-model (instead of cross-validation in stacking).
* **Use Case:** When you want faster prototyping with some risk of overfitting.

## Types of Ensemble Learning Techniques: Examples and Working

Ensemble learning techniques combine multiple models, often called **base learners** or **weak learners**, to build a stronger predictive model. The key types are **bagging**, **boosting**, **stacking**, and **voting**. Each has a distinct approach and is suited to different problems.

### 1. **Bagging (Bootstrap Aggregating)**

**How it Works:**
- Multiple models are trained independently on random subsets (with replacement) of the data.
- Their outputs are combined by averaging (for regression) or majority voting (for classification).
- Bagging primarily reduces model variance.

**Example: Random Forest**
- Consists of many decision trees, each trained on a random sample of data and features.
- The final prediction aggregates the predictions of all trees.
- Used for tasks like image classification, recognizing handwritten digits, or predicting loan defaults[1][2][3].

### 2. **Boosting**

**How it Works:**
- Models are trained sequentially; each new model focuses on correcting the errors of the previous ones.
- The final prediction is a weighted combination of all models.
- Boosting primarily reduces bias and can also reduce variance.

**Examples:**
- **AdaBoost:** Assigns higher weights to misclassified data, forcing new models to focus on the hardest samples.
- **Gradient Boosting:** Each new model tries to predict the residual (error) of the combined previous models, gradually refining the prediction.
- **XGBoost, CatBoost, LightGBM:** Optimized versions often used in competitions and industry, especially for structured/tabular data, such as fraud detection or predicting customer churn[3][4][5].

### 3. **Stacking (Stacked Generalization)**

**How it Works:**
- Multiple diverse models (e.g., decision tree, SVM, and logistic regression) are trained on the entire dataset.
- Predictions from these **base models** are themselves used as features to train a **meta-model** (or blender/stacker), which makes the final prediction.
- Stacking leverages the strengths of different model types.

**Example:**
- For a credit scoring problem, the team trains logistic regression, random forest, and an SVM. Their outputs are then used to train a meta-model—a decision tree—which outputs the final score[2][3][5].

### 4. **Voting**

**How it Works:**
- Predictions from several (possibly heterogeneous) models are combined—usually via majority (hard voting) or weighted averaging (soft voting).
- Unlike stacking, there is no meta-model.

**Example:**
- For spam detection, combine predictions from a Naive Bayes classifier, a random forest, and an SVM. If two out of three flag an email as spam, the ensemble flags it as spam[6][2].

| Boosting    | Sequential models, each corrects previous      | AdaBoost, GradientBoost| Reduces bias, high accuracy |

### Quick Illustrative Example

**Bagging Example:**
- If you have a fruit image dataset, you create several samples (with replacement).
- Train a separate decision tree on each sample.
- Predict “apple” or “orange” for a new image by majority vote from all trees.

**Boosting Example:**
- Use the entire fruit dataset, train first tree.
- Focus next tree on misclassified images—say, tricky apples.
- Repeat, each time adjusting to errors. Final output is a weighted combination of all trees.

**Stacking Example:**
- Train k-NN, SVM, and random forest on the fruit dataset.
- Use their predictions as input to a logistic regression model, which learns the best way to blend those predictions for the final output.

### Summary

- **Bagging** (Random Forest): Best for reducing variance and overfitting with parallel, independent models.
- **Boosting** (AdaBoost, Gradient Boosting): Powerful for error correction by learning sequentially and focusing on hard cases.
- **Stacking:** Combines different types of models for optimal results using a secondary learner.
- **Voting:** Aggregates decisions from multiple models simply by vote or average, improving stability and robustness.

## Difference between Bagging and Boosting

### 1. Types of models used:

Our target is to achieve Low Bias and Low Variance

- **Bagging :** We take models with Low Bias and High variance. eg: fully grown Decision Tree, KNN with low neighbor count


- **Boosting :** We take models with High Bias and Low variance. eg: decision_stump ie.e decision tree with max_depth=1

### 2. Sequential vs Parallel:

- **Bagging** is done **Parallel** as give different models a different subset of data and is trained parallely.


- **Boosting** data is given to all the models one by one with each model learning from the mistakes of previous ones.

### 3.  Weightage of base models:

- **Bagging :** Weightage of each base model is same.


- **Boosting :** Every base model has different weights. Better models have more weightage

## Voting Ensemble

The Voting classifier estimator built by combining different classification models turns out to be stronger meta-classifier that balances out the individual classifiers’ weaknesses on a particular dataset.


In Voting Ensemble for Regression: We pass the data through all the models and then take mean of all the outputs.

In Voting Ensemble for Classification: We take the most occuring output through all the models as final output.

#### Why voting works:

1. Models should be dissimilar and independent. Similar model altogether won't give good results.

2. Accuracy of each model should be minimum 50%.

Voting Classifier : https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.VotingClassifier.html
        
Voting Regressor:  https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.VotingRegressor.html

### Voting Types:

voting {‘hard’, ‘soft’}, default=’hard’


1. Hard



2. Soft




If ‘hard’, uses predicted class labels for majority rule voting. ie. 1 or 0. And based on Majority of occurence we select the output.

But in ‘soft’, predicts the class label based on the argmax of the sums of the predicted probabilities, which is recommended for an ensemble of well-calibrated classifiers. 

eg: 

M1 = 0.9 = 1 and 0.1 = 0, 

M2 = 0.8 = 1 and 0.2 = 0,

M3 = 0.7 = 1 and 0.3 = 0

so soft voting will find average of all and for final answer 1 = 0.8 and 0 = 0.2

<br></br>
Hard voting classifier classifies data based on class labels and the weights associated with each classifier.

Soft voting classifier classifies data based on the probabilities and the weights associated with each classifier.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.head()
```

```python
df.columns = iris.feature_names
```

```python
df.head()
```

```python
df['Species'] = iris.target
```

```python
df.head()
```

```python
import seaborn as sns
sns.pairplot(df,hue='Species')
```

```python
# to make classification task difficult we will take only 2 columns i.e Sepal_length and Sepal_width

new_df = df[df['Species']!=0][['sepal length (cm)','sepal width (cm)','Species']]
new_df.head()
```

```python
new_df.shape
```

```python
X = new_df.iloc[:,0:2]
X.head()
```

```python
y = new_df.iloc[:,-1] 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
clf1 = LogisticRegression()
clf2 = RandomForestClassifier()
clf3 = KNeighborsClassifier()
```

```python
# creating estimator objects should be list of tuples('string', estimator)
# har tuple mein pehla naam hai next uss algorithm ka object hai

estimators = [('lr',clf1),('rf',clf2),('knn',clf3)]
```

```python
# cross validation on each estimator
for estimator in estimators:
    x = cross_val_score(estimator[1],X,y,cv=10,scoring='accuracy',n_jobs=-1)
    print(estimator[1])
    print(estimator[0],np.round(np.mean(x),2))

# getting accuracy of each model
```

```python
from sklearn.ensemble import VotingClassifier
```

### Hard Voting 

```python
vc = VotingClassifier(estimators = estimators,voting='hard')
x = cross_val_score(vc,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))

# getting accuracy of hard voting
```

### Soft Voting

```python
vc1 = VotingClassifier(estimators = estimators, voting ='soft')
x = cross_val_score(vc1,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))

# getting accuracy of Soft voting
```

### Weighted Voting

giving different weightage to different models

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            
            vc = VotingClassifier(estimators=estimators, voting = 'soft',weights = [i,j,k]) # weights = n_classifiers
            x = cross_val_score(vc,X,y,cv=10,scoring='accuracy')
            print("for i={}, j={}, k={}".format(i,j,k),np.round(np.mean(x),2))
```

best accuracy "72%" is when i=logistic which has 75% accuracy is given more weightage and j= random_forest and k = knn are given equal weights

## Voting Regressor

```python
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
X.shape
```

```python
y.shape
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr',lr),('dt',dt),('svr',svr)]
```

```python
for estimator in estimators:
    
    scores = cross_val_score(estimator[1],X,y, scoring='r2',cv=10)
    print(estimator[0],np.round(np.mean(scores),2))
```

##### now with Voting Regressor

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators)

scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(scores),2))
```

##### experimenting with weights

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            
            vr = VotingRegressor(estimators,weights=[i,j,k])
            
            scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
            
            print("For i={}, j={}, k={}".format(i,j,k),np.round(np.mean(scores),2))
    
```

best accuracy is for i=2, j=1, k=1  = 46%

#### Method 2: Regressor for Same Algorithm

```python
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=2)
dt3 = DecisionTreeRegressor(max_depth=3)
dt4 = DecisionTreeRegressor(max_depth=4)
dt5 = DecisionTreeRegressor(max_depth=5)
```

```python
estimators = [('dt1',dt1),('dt2',dt2),('dt3',dt3),('dt4',dt4),('dt5',dt5)]
```

```python
for estimator in estimators:
    
    scores = cross_val_score(estimator[1],X,y,scoring='r2',cv=10)
    
    print(estimator[0],np.round(np.mean(scores),2))
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting regressor",np.round(np.mean(scores),3))
```
